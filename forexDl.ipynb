```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("proper.csv", delimiter="\t")
```

```python
dataset.drop("<VOL>", inplace=True, axis = 1)
```

```python
from datetime import datetime
import calendar
```

```python
dataset['<DATE>'] = dataset['<DATE>'].apply(lambda x: calendar.timegm(datetime.strptime(x, '%Y.%m.%d').utctimetuple()))
```

```python
def EMA(data, col, n, arr):
    y = n-1
    for i in range(data['<CLOSE>'].size):
        if(i < y):
            arr.append(0)
        elif(i == y):
            av = data.iloc[:i, col].mean()
            arr.append(av)
        else:
            EM = ((data.iloc[i,col] * 2)/(n + 1)) + arr[i-1] * (1-(2/(n+1)))
            arr.append(EM)
```

```python
EMA12 = []
EMA(dataset, 4, 12, EMA12)
EMA26 = []
EMA(dataset, 4, 26, EMA26)
```

```python
dataset['<EMA12>'] = EMA12
dataset['<EMA26>'] = EMA26
dataset['<MACD>'] = dataset['<EMA12>'] - dataset['<EMA26>']
```

```python
dataset = dataset.iloc[11:,:]
```

```python
SIGNAL = []
EMA(dataset, 9, 9, SIGNAL)
dataset['<SIGNAL>'] = SIGNAL
```

```python
dataset = dataset.iloc[9:,:]
```

```python
def GL(data, gains, losses, col):
    for i in range(data['<CLOSE>'].size):
        if(i == 0):
            gains.append(0)
            losses.append(0)
        else:
            diff = data.iloc[i, col] - data.iloc[i-1, col]
            if(diff > 0):
                gains.append(diff)
                losses.append(0)
            elif(diff == 0):
                gains.append(0)
                losses.append(0)
            else:
                gains.append(0)
                losses.append(diff)
```

```python
GAINS = []
LOSSES = []
GL(dataset, GAINS, LOSSES, 4)
```

```python
dataset['<GAINS>'] = GAINS
dataset['<LOSSES>'] = LOSSES
```

```python
dataset.head(20)
```

```python
dataset = dataset.iloc[1:, :]
```

```python
def RS(data, n, rs):
    y = n-1
    for i in range(data['<CLOSE>'].size):
        if(i < y):
            rs.append(0)
        else:
            GA = data.iloc[(i-y):i, 11].mean()
            LO = data.iloc[(i-y):i, 12].mean()
            div = GA/LO
            rs.append(div)
```

```python
RelStr = []
RS(dataset, 14, RelStr)
```

```python
dataset['<RS>'] = RelStr
```

```python
dataset = dataset.iloc[14:, :]
```

```python
dataset["<GROW>"] = dataset["<CLOSE>"] - dataset["<OPEN>"]
```

```python
dataset.head()
```

```python
def CLGrowth(i):
        if(i == 0):
            return "one"
        elif(i < 0):
            return "zero"
        elif(i > 0):
            return "two"
```

```python
dataset['<CLGROWTH>'] = dataset['<GROW>'].apply(CLGrowth)
```

```python
dataset.head()
```

```python
dataset["<CLGROWTH>"] = dataset["<CLGROWTH>"].shift(-1)
dataset = dataset.iloc[:-1,:]
```

```python
dataset.drop("<GROW>", inplace=True, axis = 1)
```

```python
X = dataset.iloc[: ,:-1].values
y = dataset.iloc[:, -1].values
```

```python
y
```

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
y = np.array(ct.fit_transform(y.reshape(-1,1)))
```

```python
y
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 0:13])
X[:, 0:13] = imputer.transform(X[:, 0:13])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_Xtrain = StandardScaler()
X_train = sc_Xtrain.fit_transform(X_train)

sc_Xtest = StandardScaler()
X_test = sc_Xtest.fit_transform(X_test)
```

```python
X_train
```

```python
import tensorflow as tf
```

```python
ann = tf.keras.models.Sequential()
```

```python
ann.add(tf.keras.layers.Dense(units=12, activation='relu'))
ann.add(tf.keras.layers.Dense(units=12, activation='relu'))
```

```python
ann.add(tf.keras.layers.Dense(units=3, activation='softmax'))
```

```python
ann.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'] )
```

```python
ann.fit(X_train, y_train, batch_size=32, epochs= 200)
```

```python
y_pred = ann.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
yp = y_pred.argmax(1)
```

```python
yt = y_test.argmax(1)
```

```python
cmT = confusion_matrix(yt, yp)
accT = accuracy_score(yt, yp)
```

```python
cmT
```

```python
accT
```
